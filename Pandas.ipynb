```python
## Pandas Rank
# importing pandas as pd
import pandas as pd

# Creating the Series
sr = pd.Series([10, 25, 3, 11, 24, 6, 25, 26])
  
# Create the Index
index_ = ['Coca Cola', 'Sprite', 'Coke', 'Fanta', 'Dew', 'ThumbsUp', 'Appy', 'fruity']
  
# set the index
sr.index = index_
  
# Print the series
print(sr)
```

```python
# assign rank
result = sr.rank(method="dense")

print(result)
```

```python
# importing pandas package
import pandas as pd
  
# making data frame from csv file
data = pd.read_csv("nba.csv")
  
# creating a rank column and passing the returned rank series
data["Rank"] = data["Age"].rank(method="dense")
  
# display
data
  
# sorting w.r.t name column
data.sort_values("Rank", inplace = True)
  
# # display after sorting w.r.t Name column
data
```
